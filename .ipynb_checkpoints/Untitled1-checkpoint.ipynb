```python
# Import necessary libraries
import qlib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from qlib.data import D
from qlib.workflow import R
from qlib.contrib.model.pytorch_gru import GRUModel  # ✅ Corrected Import
from qlib.contrib.strategy.signal_strategy import TopkDropoutStrategy
from qlib.contrib.evaluate import risk_analysis
from qlib.utils import init_instance_by_config
from qlib.workflow.task.utils import list_recorders

# Initialize Qlib
qlib.init(provider_uri="~/.qlib/qlib_data/cn_data", region="cn")

print("✅ Qlib successfully initialized!")
```

```python
# Load stock data (ensure this file exists)
file_path = "indian_stocks_last_2_years.csv"  # Update with the correct path if needed
df = pd.read_csv(file_path)

# Convert date column to datetime & sort values
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(by=['ticker', 'date'])

# Display dataset info
print("Dataset Shape:", df.shape)
print(df.head())
```

```python
# Plot stock price trends
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='close', hue='ticker')
plt.title("Stock Price Trends")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend(title='Stock')
plt.show()
```

```python
import qlib
from qlib.config import REG_CN

# Initialize Qlib (VERY IMPORTANT)
qlib.init(provider_uri="~/.qlib/qlib_data/cn_data", region=REG_CN, auto_reload=True)
```

```python
# Convert dataset to Qlib-compatible format
def convert_to_qlib_format(df):
    df = df.rename(columns={'date': 'datetime', 'close': 'feature.close', 'open': 'feature.open',
                             'high': 'feature.high', 'low': 'feature.low', 'volume': 'feature.volume'})
    df = df[['datetime', 'ticker', 'feature.close', 'feature.open', 'feature.high', 'feature.low', 'feature.volume']]
    return df

df_qlib = convert_to_qlib_format(df)
print("✅ Data converted to Qlib format!")
```
