```python
import yfinance as yf
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import mplfinance as mpf
from datetime import datetime, timedelta
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from matplotlib.widgets import Cursor

# Define available companies
available_companies = ['RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'ICICIBANK.NS', 'INFY.NS',
                     'HINDUNILVR.NS', 'SBIN.NS', 'BHARTIARTL.NS', 'ITC.NS', 'LT.NS']

# Get company selection
print("\nAvailable companies:")
print(", ".join(available_companies))
selected_company = input("Enter company symbol (e.g. RELIANCE.NS): ").strip().upper()
while selected_company not in available_companies:
    print("Invalid company. Please choose from the list.")
    selected_company = input("Enter company symbol: ").strip().upper()

# Get chart type selection
print("\nAvailable chart types: line, scatter, bar, histogram")
chart_type = input("Enter chart type: ").strip().lower()
while chart_type not in ['line', 'scatter', 'bar', 'histogram']:
    print("Invalid chart type. Please choose from: line, scatter, bar, histogram")
    chart_type = input("Enter chart type: ").strip().lower()

# Fixed training period (2 years)
start_date = datetime(2022, 1, 1)
end_date = datetime(2024, 1, 1)

# Load data
ticker = yf.Ticker(selected_company)
data = ticker.history(start=start_date, end=end_date)

# Convert index to DatetimeIndex if it's not already
if not isinstance(data.index, pd.DatetimeIndex):
    data.index = pd.to_datetime(data.index)

# Calculate indicators
def calculate_indicators(df):
    df['SMA_50'] = df['Close'].rolling(window=50).mean()
    df['SMA_200'] = df['Close'].rolling(window=200).mean()
    df['EMA_50'] = df['Close'].ewm(span=50, adjust=False).mean()
    df['EMA_200'] = df['Close'].ewm(span=200, adjust=False).mean()
    return df.dropna()

data = calculate_indicators(data)

# Display candlestick chart
mpf.plot(data, type='candle', style='charles',
        title=f'{selected_company} Historical Data (2022-01-01 to 2024-01-01)',
        ylabel='Price (₹)', volume=True, figratio=(12,6))

# Preprocess data
def preprocess_data(df):
    prices = df['Close'].values
    mean, std = prices.mean(), prices.std()
    normalized = (prices - mean) / std
    
    seq_len = 60  # Sequence length
    X, y = [], []
    for i in range(len(normalized) - seq_len):
        X.append(normalized[i:i+seq_len])
        y.append(normalized[i+seq_len])
    
    X = torch.tensor(np.array(X), dtype=torch.float32).unsqueeze(-1)  # [samples, seq_len, features]
    y = torch.tensor(np.array(y), dtype=torch.float32)
    
    return X, y, mean, std

X, y, price_mean, price_std = preprocess_data(data)

# LSTM Model
class PricePredictor(nn.Module):
    def __init__(self, input_size=1, hidden_size=64, num_layers=2):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True)
        self.linear = nn.Linear(hidden_size, 1)
        
    def forward(self, x):
        lstm_out, _ = self.lstm(x)
        return self.linear(lstm_out[:, -1])

model = PricePredictor()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.MSELoss()

# Training with early stopping
best_loss = float('inf')
patience = 20
no_improve = 0
losses = []

for epoch in range(500):
    model.train()
    optimizer.zero_grad()
    outputs = model(X)
    loss = criterion(outputs.squeeze(), y)
    loss.backward()
    optimizer.step()
    
    losses.append(loss.item())
    
    if loss.item() < best_loss:
        best_loss = loss.item()
        no_improve = 0
        torch.save(model.state_dict(), 'best_model.pth')
    else:
        no_improve += 1
        if no_improve >= patience:
            print(f"Early stopping at epoch {epoch+1}")
            break
            
    if epoch % 50 == 0:
        print(f"Epoch {epoch+1}, Loss: {loss.item():.6f}")

# Plot training loss
plt.figure(figsize=(10, 4))
plt.plot(losses)
plt.title('Training Loss Curve')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.grid(True)
plt.show()

model.load_state_dict(torch.load('best_model.pth'))

# Generate predictions
model.eval()
future_days = 30
predictions = []
current_seq = X[-1:].clone()  # shape: [1, seq_len, 1]

with torch.no_grad():
    for _ in range(future_days):
        pred = model(current_seq)  # shape: [1, 1]
        predictions.append(pred.item())
        
        # Reshape prediction to match current_seq dimensions
        new_pred = pred.reshape(1, 1, 1)  # shape: [1, 1, 1]
        
        # Update sequence by removing oldest and adding new prediction
        current_seq = torch.cat([current_seq[:, 1:, :], new_pred], dim=1)

# Denormalize predictions
predictions = (np.array(predictions) * price_std) + price_mean

# Generate future dates (business days only)
pred_dates = []
current_date = end_date + timedelta(days=1)
while len(pred_dates) < future_days:
    if current_date.weekday() < 5:  # Monday-Friday
        pred_dates.append(current_date)
    current_date += timedelta(days=1)

pred_df = pd.DataFrame({
    'Date': pred_dates,
    'Predicted Price': predictions
})

# Evaluation metrics
with torch.no_grad():
    y_pred = model(X).squeeze().numpy()
    y_true = y.numpy()

mse = mean_squared_error(y_true, y_pred)
mae = mean_absolute_error(y_true, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_true, y_pred)

print("\nEvaluation Metrics:")
print(f"MSE: {mse:.4f}")
print(f"MAE: {mae:.4f}")
print(f"RMSE: {rmse:.4f}")
print(f"R² Score: {r2:.4f}")

# Plot actual vs predicted
plt.figure(figsize=(12, 6))
plt.plot(y_true, label='Actual Prices', color='blue')
plt.plot(y_pred, label='Predicted Prices', color='orange')
plt.title('Actual vs Predicted Prices (Normalized)')
plt.xlabel('Days')
plt.ylabel('Normalized Price')
plt.legend()
plt.grid(True)
plt.show()

# Create figure for predictions plot
fig, ax = plt.subplots(figsize=(12, 6))

# Plot based on chart type
if chart_type == 'line':
    line, = ax.plot(pred_df['Date'], pred_df['Predicted Price'], 'go-', 
                   markersize=8, linewidth=2, markerfacecolor='white')
elif chart_type == 'scatter':
    scatter = ax.scatter(pred_df['Date'], pred_df['Predicted Price'], 
                        c='green', s=100, alpha=0.7)
elif chart_type == 'bar':
    bars = ax.bar(pred_df['Date'], pred_df['Predicted Price'], color='green')
elif chart_type == 'histogram':
    n, bins, patches = ax.hist(pred_df['Predicted Price'], bins=10, color='green', alpha=0.7)
    ax.set_xlabel('Predicted Price (₹)')
    ax.set_ylabel('Frequency')

# Add hover functionality for all chart types
annot = ax.annotate("", xy=(0,0), xytext=(20,20), textcoords="offset points",
                   bbox=dict(boxstyle="round", fc="w", alpha=0.9),
                   arrowprops=dict(arrowstyle="->"))
annot.set_visible(False)

def update_annot(ind, chart_type):
    if chart_type == 'line':
        x, y = line.get_data()
        annot.xy = (x[ind["ind"][0]], y[ind["ind"][0]])
        date_str = pred_df['Date'].iloc[ind["ind"][0]].strftime('%b %d, %Y')
        price = pred_df['Predicted Price'].iloc[ind["ind"][0]]
        annot.set_text(f"Date: {date_str}\nPrice: ₹{price:.2f}")
    elif chart_type == 'scatter':
        pos = scatter.get_offsets()[ind["ind"][0]]
        annot.xy = (pos[0], pos[1])
        date_str = pred_df['Date'].iloc[ind["ind"][0]].strftime('%b %d, %Y')
        price = pred_df['Predicted Price'].iloc[ind["ind"][0]]
        annot.set_text(f"Date: {date_str}\nPrice: ₹{price:.2f}")
    elif chart_type == 'bar':
        bar = bars[ind["ind"][0]]
        annot.xy = (bar.get_x() + bar.get_width()/2, bar.get_height())
        date_str = pred_df['Date'].iloc[ind["ind"][0]].strftime('%b %d, %Y')
        price = pred_df['Predicted Price'].iloc[ind["ind"][0]]
        annot.set_text(f"Date: {date_str}\nPrice: ₹{price:.2f}")
    elif chart_type == 'histogram':
        # For histogram, show the bin range and count
        bin_idx = np.searchsorted(bins, ind["ind"][0], side='right') - 1
        bin_range = f"{bins[bin_idx]:.2f}-{bins[bin_idx+1]:.2f}"
        count = n[bin_idx]
        annot.xy = (bins[bin_idx] + (bins[bin_idx+1]-bins[bin_idx])/2, count)
        annot.set_text(f"Price Range: ₹{bin_range}\nCount: {int(count)}")

def hover(event):
    vis = annot.get_visible()
    if event.inaxes == ax:
        if chart_type == 'line':
            cont, ind = line.contains(event)
        elif chart_type == 'scatter':
            cont, ind = scatter.contains(event)
        elif chart_type == 'bar':
            cont, ind = bars.contains(event)
        elif chart_type == 'histogram':
            # For histogram, find which bin the mouse is in
            if len(bins) > 0:
                x = event.xdata
                if x >= bins[0] and x <= bins[-1]:
                    bin_idx = np.searchsorted(bins, x, side='right') - 1
                    if bin_idx >= 0 and bin_idx < len(n):
                        ind = {"ind": [bin_idx]}
                        cont = True
                    else:
                        cont = False
                else:
                    cont = False
            else:
                cont = False
        
        if cont:
            update_annot(ind, chart_type)
            annot.set_visible(True)
            fig.canvas.draw_idle()
        else:
            if vis:
                annot.set_visible(False)
                fig.canvas.draw_idle()

fig.canvas.mpl_connect("motion_notify_event", hover)

# Add cursor for all chart types
cursor = Cursor(ax, horizOn=True, vertOn=True, color='red', linewidth=1)

# Format the plot
ax.set_title(f'{selected_company} - 30-Day Price Prediction ({chart_type.capitalize()})')
if chart_type != 'histogram':
    ax.set_xlabel('Date')
    ax.set_ylabel('Predicted Price (₹)')
    plt.xticks(rotation=45)
else:
    ax.set_xlabel('Predicted Price (₹)')
    ax.set_ylabel('Frequency')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Print predictions
print("\n30-Day Price Predictions:")
print(pred_df.round(2))

# Save predictions
pred_df.to_csv(f'{selected_company}_predictions.csv', index=False)
print(f"\nPredictions saved to {selected_company}_predictions.csv")
```
