```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mplfinance as mpf
import yfinance as yf
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import mean_absolute_error


# Define the companies
companies = ['RELIANCE.NS', 'TCS.NS', 'HDFCBANK.NS', 'ICICIBANK.NS', 'INFY.NS',
             'HINDUNILVR.NS', 'SBIN.NS', 'BHARTIARTL.NS', 'ITC.NS', 'LT.NS']

# Display company names for user selection
print("Select a company from the list below:")
for i, company in enumerate(companies):
    print(f"{i + 1}. {company}")

# Get user input
while True:
    choice = input("Enter the number of your chosen company: ")
    try:
        choice = int(choice)
        if 1 <= choice <= len(companies):
            selected_company = companies[choice - 1]
            break
        else:
            print("Invalid choice. Please enter a number within the range.")
    except ValueError:
        print("Invalid input. Please enter a number.")


# Load data for the selected company
ticker = yf.Ticker(selected_company)
data = ticker.history(period='2y')

# Preprocess data for prediction
def preprocess_data(data):
    close_prices = pd.to_numeric(data['Close'], errors='coerce').dropna().values

    if len(close_prices) == 0:
        return None, None, None, None

    price_mean = close_prices.mean()
    price_std = close_prices.std()
    normalized_prices = (close_prices - price_mean) / price_std

    seq_len = 100
    sequences = []
    labels = []
    
    for i in range(len(normalized_prices) - seq_len):
        seq = normalized_prices[i:i + seq_len]
        label = normalized_prices[i + seq_len]
        sequences.append(seq)
        labels.append(label)

    sequences = np.array(sequences)
    labels = np.array(labels)

    sequences = torch.tensor(sequences, dtype=torch.float32).unsqueeze(-1)
    labels = torch.tensor(labels, dtype=torch.float32)

    return sequences, labels, price_mean, price_std

# Apply function
sequences, labels, price_mean, price_std = preprocess_data(data)

# Train-test split (80% train, 20% test)
train_size = int(len(sequences) * 0.8)
train_sequences, test_sequences = sequences[:train_size], sequences[train_size:]
train_labels, test_labels = labels[:train_size], labels[train_size:]


# Function to calculate indicators
def calculate_indicators(data):
    # Calculate Simple Moving Average (SMA)
    data['SMA_50'] = data['Close'].rolling(window=50).mean()
    data['SMA_200'] = data['Close'].rolling(window=200).mean()

    # Calculate Exponential Moving Average (EMA)
    data['EMA_50'] = data['Close'].ewm(span=50, adjust=False).mean()
    data['EMA_200'] = data['Close'].ewm(span=200, adjust=False).mean()

    # Calculate Rate of Change (ROC)
    data['ROC'] = ((data['Close'] - data['Close'].shift(14)) / data['Close'].shift(14)) * 100

    # Calculate Average True Range (ATR)
    data['High-Low'] = data['High'] - data['Low']
    data['High-Close'] = abs(data['High'] - data['Close'].shift(1))
    data['Low-Close'] = abs(data['Low'] - data['Close'].shift(1))
    data['True_Range'] = data[['High-Low', 'High-Close', 'Low-Close']].max(axis=1)
    data['ATR'] = data['True_Range'].rolling(window=14).mean()

    # Calculate On-Balance Volume (OBV)
    data['OBV'] = np.where(data['Close'] > data['Close'].shift(1), data['Volume'], -data['Volume']).cumsum()
    
    return data

# Get user input for indicator category
print("Select an indicator category:")
indicator_categories = {
    1: 'Trend-Following',
    2: 'Momentum',
    3: 'Volatility',
    4: 'Volume'
}

for key, value in indicator_categories.items():
    print(f"{key}. {value}")

while True:
    try:
        category_choice = int(input("Enter the number for your chosen indicator category: "))
        if category_choice in indicator_categories:
            indicator_category = indicator_categories[category_choice].lower()
            break
        else:
            print("Invalid choice. Please select a valid option.")
    except ValueError:
        print("Invalid input. Please enter a number.")


data = calculate_indicators(data)  # Assuming 'data' is already loaded

# Plot candlesticks
mpf.plot(data, type='candle', title=f'{selected_company} Candlestick Chart', ylabel='Price', ylabel_lower='Volume', volume=True)

# Plot selected indicators
plt.figure(figsize=(12, 8))

if indicator_category == 'trend-following':
    plt.subplot(2, 1, 1)
    plt.plot(data['Close'], label='Close Price')
    plt.plot(data['SMA_50'], label='SMA 50')
    plt.plot(data['SMA_200'], label='SMA 200')
    plt.plot(data['EMA_50'], label='EMA 50')
    plt.plot(data['EMA_200'], label='EMA 200')
    plt.legend(loc='upper left')
    plt.title('Trend-Following Indicators (SMA & EMA)')

elif indicator_category == 'momentum':
    plt.plot(data['ROC'], label='ROC')
    plt.axhline(y=0, color='r', linestyle='--')
    plt.legend(loc='upper left')
    plt.title('Momentum Indicator (ROC)')

elif indicator_category == 'volatility':
    plt.plot(data['ATR'], label='ATR')
    plt.legend(loc='upper left')
    plt.title('Volatility Indicator (ATR)')

elif indicator_category == 'volume':
    plt.plot(data['OBV'], label='OBV')
    plt.legend(loc='upper left')
    plt.title('Volume Indicator (OBV)')

else:
    print("Invalid selection. Please choose a valid indicator category.")

plt.tight_layout()
plt.show()
```

```python
# Define the SFM Cell
class SFMCell(nn.Module):
    def __init__(self, input_dim, hidden_dim, freq_dim):  # Fix here
        super(SFMCell, self).__init__()  # Fix here
        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.freq_dim = freq_dim

        self.state_gate = nn.Linear(input_dim + hidden_dim, hidden_dim)
        self.freq_gate = nn.Linear(input_dim + hidden_dim, hidden_dim)
        self.output_layer = nn.Linear(hidden_dim, 1)

    def forward(self, x, prev_state):
        state_gate_output = torch.sigmoid(self.state_gate(torch.cat((x, prev_state), dim=1)))
        freq_gate_output = torch.sigmoid(self.freq_gate(torch.cat((x, prev_state), dim=1)))

        new_state = state_gate_output * prev_state + (1 - state_gate_output) * freq_gate_output
        output = self.output_layer(new_state)

        return output, new_state

# Define the SFM Model
class SFMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, freq_dim):  # Fix here
        super(SFMModel, self).__init__()  # Fix here
        self.sfm_cell = SFMCell(input_dim, hidden_dim, freq_dim)

    def forward(self, x, initial_state):
        outputs = []
        state = initial_state

        for t in range(x.size(1)):
            output, state = self.sfm_cell(x[:, t].unsqueeze(-1), state)
            outputs.append(output)

        outputs = torch.stack(outputs).transpose(0, 1)
        return outputs
```
